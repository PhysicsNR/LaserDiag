```python
# LaserDiag Demo Notebook

import sys, os
sys.path.append(os.path.abspath('../'))

from core.io import load_csv
from core.analysis import fft_analysis, thd_calc
from core.plot import plot_fft

import numpy as np

# Simulate data: sine wave + harmonics + noise
fs = 10000  # 10 kHz sample rate
t = np.arange(0, 1, 1/fs)
signal = 1.0 * np.sin(2 * np.pi * 1000 * t) + 0.2 * np.sin(2 * np.pi * 2000 * t) + 0.1 * np.random.randn(len(t))

# FFT Analysis
f, spectrum = fft_analysis(signal, fs)
plot_fft(f, spectrum, filename="output_plots/fft_example.png")

# THD Calculation
thd = thd_calc(signal, fs, f0=1000)
print("Total Harmonic Distortion:", thd)
```
